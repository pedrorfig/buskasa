```python
import pandas as pd
```

```python
raw_matches = pd.read_csv(r"C:\Users\58781\OneDrive - Bain\Desktop\matches.csv", index_col=[0,1])
listings = pd.read_csv(r"C:\Users\58781\OneDrive - Bain\Desktop\listings_to_add.csv", index_col=[1]).drop(columns=['Unnamed: 0'])
display(raw_matches)
display(listings)
```

```python
matches = raw_matches.copy()
```

```python
matches['similarity'] = matches.sum(axis=1) / len(matches.columns)
matches['max_similarity'] = matches.groupby(level=1)['similarity'].transform('max')
matches = matches[matches['similarity'] == matches['max_similarity']]
matches
```

```python
mask = ~matches.index.get_level_values(1).duplicated(keep='first')
matches = matches[mask]
matches
```

```python
mask = ~matches.index.get_level_values(1).duplicated(keep='first')
matches = matches[mask]
matches
```

```python

matches = matches.drop(columns=matches.columns)
level_0_rows = [(level0_value, level0_value) for level0_value in 
                matches.index.get_level_values('listing_id_1').unique()]
new_rows_matches = pd.MultiIndex.from_tuples(level_0_rows)
matches = matches.index.union(new_rows_matches)
matches.names = ('listing_id_1', 'listing_id')   
matches = matches.to_frame().reset_index(drop=True)
matches
```

```python
merged_df = matches.merge(listings, left_on='listing_id', right_index=True).set_index(['listing_id_1', 'listing_id'])
merged_df
```

```python
# Leave only cheapest listing of the dupliated
cheapest_listings_index = merged_df.groupby(level=0)['price'].idxmin()
cheapest_listings_index
```

```python
# merged_df.index = merged_df.index.droplevel(0)
cheapest_listings = merged_df.loc[cheapest_listings_index]
cheapest_listings.index = cheapest_listings.index.droplevel(0)
cheapest_listings
```

```python
listings_dropped_all_matches = listings.drop(matches['listing_id'].unique())
result = pd.concat([cheapest_listings, listings_dropped_all_matches])
result
```
