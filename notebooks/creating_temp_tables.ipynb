```python
import uuid

import pandas as pd
from etl_modules import extract
```

```python
_engine = extract.create_db_engine()
```

```python
temp_table_name = f"temp_{uuid.uuid4().hex[:6]}"
```

```python

with _engine.connect() as conn:
    data = pd.read_sql(
        """
        SELECT *
        FROM listings
        LIMIT 20
        """, con=conn)
    data.to_sql(con=_engine, if_exists="replace", schema='public', name=temp_table_name)
    
data
```
